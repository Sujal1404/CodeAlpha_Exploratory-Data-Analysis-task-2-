#  Data Analytics Project – Task 2: Exploratory Data Analysis (EDA)

### **Objective**
To perform Exploratory Data Analysis (EDA) on the scraped dataset to understand its structure, detect trends, patterns, anomalies, and prepare it for further analysis.

## **Step 1: Import Required Libraries**

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For better visualization
sns.set(style="whitegrid")
```

## **Step 2: Load Dataset**

```python

# Load dataset (replace with path if needed)
df = pd.read_csv("quotes_dataset.csv")
df.head()
```

## **Step 3: Explore Data Structure**

```python

# Shape of dataset
print("Shape of dataset:", df.shape)

# Data types and info
df.info()

# Check for missing values
df.isnull().sum()
```

## **Step 4: Basic Statistics & Unique Values**

```python

# Unique authors count
print("Number of unique authors:", df['Author'].nunique())

# Top 10 authors with most quotes
top_authors = df['Author'].value_counts().head(10)
top_authors
```

## **Step 5: Visualization – Top Authors by Quotes**

```python

plt.figure(figsize=(10,6))
sns.barplot(x=top_authors.values, y=top_authors.index, palette="viridis")
plt.title("Top Authors by Number of Quotes")
plt.xlabel("Number of Quotes")
plt.ylabel("Authors")
plt.show()
```

## **Step 6: Tag Analysis**

```python

# Split tags into list
all_tags = df['Tags'].str.split(", ")

# Flatten the list
all_tags_flat = [tag for sublist in all_tags.dropna() for tag in sublist]

# Create DataFrame for tags
tags_df = pd.Series(all_tags_flat).value_counts().head(10)

tags_df
```

## **Step 7: Visualization – Top Tags**

```python

plt.figure(figsize=(10,6))
sns.barplot(x=tags_df.values, y=tags_df.index, palette="mako")
plt.title("Top 10 Tags in Quotes")
plt.xlabel("Frequency")
plt.ylabel("Tags")
plt.show()
```
